### Análise de **Regressão**

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

plt.style.use('ggplot')

df = pd.read_csv('https://raw.githubusercontent.com/gabrielluizone/Statistical-Report/main/data/regressao_Regiao_5.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="RdBu", fmt=".2f", linewidths=0)
plt.title("Gráfico de Correlação das Variáveis")
plt.show()
```

```python
X = df[['idade', 'uti', 'suportvent']]
X = sm.add_constant(X)

# Defina sua variável dependente
y = df['evolucao']

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Defina suas variáveis independentes (preditoras)
X = df[['idade', 'uti', 'suportvent']]

# Adicione uma constante para o termo de interceptação
X = sm.add_constant(X)

# Defina sua variável dependente
y = df['evolucao']

# Crie um modelo de regressão linear
model = sm.OLS(y, X).fit()

# Crie gráficos de dispersão parciais
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Gráfico de dispersão parcial para idade
axs[0].scatter(df['idade'], model.resid, alpha=0.5)
axs[0].set_xlabel('Idade')
axs[0].set_ylabel('Resíduos')
axs[0].set_title('Gráfico de Dispersão Parcial para Idade')

# Gráfico de dispersão parcial para UTI
axs[1].scatter(df['uti'], model.resid, alpha=0.5)
axs[1].set_xlabel('UTI')
axs[1].set_ylabel('Resíduos')
axs[1].set_title('Gráfico de Dispersão Parcial para UTI')

# Gráfico de dispersão parcial para suporte ventilatório
axs[2].scatter(df['suportvent'], model.resid, alpha=0.5)
axs[2].set_xlabel('Suporte Ventilatório')
axs[2].set_ylabel('Resíduos')
axs[2].set_title('Gráfico de Dispersão Parcial para Suporte Ventilatório')

plt.tight_layout()
plt.show()
```
