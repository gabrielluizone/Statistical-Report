```python
import numpy as np
from google.colab import autoviz
plt.style.use('fivethirtyeight')

def violin_plot(df, value_colname, facet_colname, figscale=1, mpl_palette_name='Paired', **kwargs):
  from matplotlib import pyplot as plt
  import seaborn as sns
  figsize = (9 * figscale, 1.2 * figscale * len(df[facet_colname].unique()))
  plt.figure(figsize=figsize)
  sns.violinplot(df, x=value_colname, y=facet_colname)
  sns.despine(top=True, right=True, bottom=True, left=True)
  return autoviz.MplChart.from_current_mpl_state()

chart = violin_plot(df, *['idade', 'evolucao'], **{'inner': 'box'})
chart
```

```python
# Função para criar gráficos de barras separados
def multi_analise(dados, lista_colunas, size=(14,6)):
    valores_coluna1 = dados[lista_colunas[0]].unique()
    num_valores_coluna1 = len(valores_coluna1)

    fig, axs = plt.subplots(1, num_valores_coluna1, figsize=size)

    for i, valor in enumerate(valores_coluna1):
        sub_dados = dados[dados[lista_colunas[0]] == valor]
        tb = sub_dados.groupby(lista_colunas[1:]).size().unstack(fill_value=0)
        ax = axs[i]

        tb.plot(kind='bar', stacked=True, ax=ax)
        ax.set_xlabel(valor)
        ax.set_title(valor)
        ax.legend(title=lista_colunas[2], bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.suptitle(f'Análise | {lista_colunas[1]} e {lista_colunas[2]} para {lista_colunas[0]}', fontsize=20)
    plt.tight_layout()
    plt.show()

    return dados.groupby(lista_colunas).size().unstack(fill_value=0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def create_stacked_bar_plot(data, x_col, y_col, figsize=(7, 4), palette='coolwarm', **kwargs):
    plt.figure(figsize=figsize)
    sns.countplot(data=data, y=x_col, hue=y_col, palette=palette, **kwargs)
    sns.despine(top=True, right=True, bottom=True, left=True)
    plt.xlabel(x_col)
    plt.ylabel(None)
    plt.title(f'Relação entre {x_col} e {y_col}')
    plt.legend(title=y_col, bbox_to_anchor=(1.05, 1), loc='upper center')
    plt.show()

# Exemplo de uso:
create_stacked_bar_plot(df, 'sexo', 'covid')
```
