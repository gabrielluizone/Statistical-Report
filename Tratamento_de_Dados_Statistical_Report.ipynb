## Tratamento de Dados
> Relatório Estatístico

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/covid_Regiao_5.xlsx')
```

### Convertendo o Sexo

```python
df['sexo'] = df['sexo'].apply(lambda x: 0 if x == 'M' else 1)
```

### Convertendo os Caracteres "NA" para NaN

```python
df = df.apply(lambda x: x.replace('NA', np.nan))
df.head(3)
```

### Convertendo as Possíveis Colunas em Numérico

```python
for coluna in df.columns:
  df[coluna] = df[coluna].astype(pd.Int64Dtype(), errors='ignore')
df.sample(3)
```
